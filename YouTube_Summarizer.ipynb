Extractive Text Summary

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
import re
```

```python
# Removing Square Brackets and Extra Spaces
text = re.sub(r'\[[0-9]*\]', ' ', text1)
text = re.sub(r'\s+', ' ', text)
text=re.sub(",", "", text)
text
```

```python
sentence_list = nltk.sent_tokenize(text)
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
words = word_tokenize(text)
words
```

```python
word_frequencies = {}
# record how many times each word will appear in the text after removing the stop words.
```

```python
for word in words:
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1
```

```python
maximum_frequncy = max(word_frequencies.values())

for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
```

```python
sentence_scores = {}
for sent in sentence_list:
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' ')) < 30:
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
sentence_scores
```

```python
import heapq
select_length=int(len(sentence_scores)*0.3)
summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

summary = ' '.join(summary_sentences)
print(summary)
print("Text:",len(text))
print("Summary",len(summary))
```

TF_IDF

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize
```

```python
stop_words = set(stopwords.words("english"))
```

```python
text=text1
```

```python
# Removing Square Brackets and Extra Spaces
text = re.sub(r'\[[0-9]*\]', ' ', text1)
text = re.sub(r'\s+', ' ', text)
text=re.sub(",", "", text)
text
```

```python
# Create the document-term matrix
sentences = sent_tokenize(text)
preprocessed_sentences = [
    " ".join(word for word in sent.lower().split() if word not in stop_words)
    for sent in sentences
]

count_vectorizer = CountVectorizer()
X = count_vectorizer.fit_transform(preprocessed_sentences)
```

```python
# Calculate TF-IDF values
tfidf_transformer = TfidfTransformer()
tfidf_matrix = tfidf_transformer.fit_transform(X)

# Summarization based on TF-IDF scores
sentence_scores = tfidf_matrix.sum(axis=1)
sentence_scores = sentence_scores.flatten()
sentence_ranking = sorted(((score, sent) for score, sent in zip(sentence_scores, sentences)), reverse=True)

# Generate summary
max_summary_sentences = 3
summary_sentences = [sent for _, sent in sentence_ranking[:max_summary_sentences]]
summary = " ".join(summary_sentences)
print(summary)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelWithLMHead
```

```python
tokenizer=AutoTokenizer.from_pretrained('T5-base')
model=AutoModelWithLMHead.from_pretrained('T5-base', return_dict=True)
```

```python
inputs=tokenizer.encode("sumarize: " +text1,return_tensors='pt', max_length=512, truncation=True)
```

```python
output = model.generate(inputs, min_length=1500, max_length=5000)
```

```python
def preprocess_sentence(sentence):
    # Remove <pad> type of substrings
    cleaned_sentence = re.sub(r'<[^>]+>', '', sentence)
    # Remove anything other than words and punctuation
    cleaned_sentence = re.sub(r'[^a-zA-Z\s.,?!]', '', cleaned_sentence)
    return cleaned_sentence
```

```python
summary=tokenizer.decode(output[0])
cleaned_sentence = preprocess_sentence(summary)
print(cleaned_sentence)
```
